```python
def E_cuts(energy,L1,L2,ra,dec):
    idx_arr = np.logical_and(energy>=L1,energy<=L2)
    return idx_arr,energy[idx_arr],ra[idx_arr],dec[idx_arr] 
```

```python
def plot_RA_DEC_for_Ecuts(energy,L1,L2,ra,dec,colour=True):
    idx, en, ra, dec=E_cuts(energy,L1,L2,ra,dec)
    df = pd.DataFrame({'x':24*(ra/360), 'y':dec})
    g = sns.jointplot(x='x', y='y', data=df, kind = 'hex',
                  stat_func=None,
                  color=colour,
                  marginal_kws=dict(bins=N_bins))
    g.set_axis_labels("RA", "DEC");
    #g.fig.subplots_adjust(top=0.05,bottom=0.01)
    g.fig.suptitle('%d-%d MeV'%(L1,L2), y=1.001)
    plot_tools.tick_par(plt)
    plot_tools.plot_size(plt,mul=1) 
    plt.tight_layout()
    plt.savefig('%s/RA_vs_DEC_%s-%s.pdf'%(res_dir,str(L1),str(L2)))
```

```python
plot_RA_DEC_for_Ecuts(Energy,100,300,RA,DEC,"#9b59b6")
plot_RA_DEC_for_Ecuts(Energy,300,1000,RA,DEC,"#e74c3c")
plot_RA_DEC_for_Ecuts(Energy,1000,3000,RA,DEC,"#34495e")
plot_RA_DEC_for_Ecuts(Energy,3000,5000,RA,DEC,"#2ecc71")
```

```python
def plot_hist_err(x,probs,lb):
    y, bin_edges = numpy.histogram(x,bins=100, 
         weights = probs)
    yerr = np.sqrt(y)
    plt.errorbar((bin_edges[:-1]+bin_edges[1:])/2, y, yerr,
                     drawstyle='steps-mid',
                     capsize=1,label=lb)
    
```

```python
E_r = np.array([100,300,1000,5000])

plot_hist_err(wpt.Fermi_MET_to_MJD(Time),Ph_prob,'No\_E\_cuts')
plt.axvline(mjd_disappear, color = 'grey')

for i in range(len(E_r)-1):
    idx, en , ra, dec = E_cuts(Energy,E_r[i],E_r[i+1],RA,DEC)
    Time_E_cuts = Time[idx]
    Ph_prob_E_cuts = Ph_prob[idx]
    lb = "%d-%d MeV"%(E_r[i],E_r[i+1])
    plot_hist_err(wpt.Fermi_MET_to_MJD(Time_E_cuts),
                  Ph_prob_E_cuts,lb)

plt.legend(loc='best',fontsize=8)
plot_tools.tick_par(plt)
plt.xlabel('MJD')
plt.ylim(-25,200)
plt.xlim(54600,57800)
plt.ylabel('Prob\_weighted\_photons')
plt.gcf().set_size_inches(8,3.5)
plt.tight_layout()
plt.savefig('%s/Lightcurves_photons_w_E_cuts.pdf'%res_dir)
```
